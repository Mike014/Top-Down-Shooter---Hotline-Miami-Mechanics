# Struttura del Progetto: Top-Down Shooter

## Enemy_Scripts/

| File | Descrizione |
| --- | --- |
| **EnemiesAnimationHandler.cs** | Gestisce le animazioni dei nemici (movimento, danno, morte). Imposta i parametri `vSpeed`/`hSpeed` nell'Animator e controlla i bool `isDamaged`/`isDead`. |
| **EnemyDrop.cs** | Sistema di drop per nemici. Quando un nemico muore, c'è una probabilità (default 10%) di droppare un'arma casuale dall'array assegnato. |
| **EnemySpawner.cs** | Spawna nemici casuali attorno al player a intervalli regolari (default 2s), a una distanza fissa con piccola variazione casuale. |

### Enemy_Scripts/Mover/

| File | Descrizione |
| --- | --- |
| **BatMover.cs** | Nemico "Bat" che insegue il player. Infligge danno a contatto, esegue animazione di morte e può droppare oggetti. Reagisce anche ai proiettili. |
| **BeybladeMover.cs** | Nemico che segue un percorso di checkpoint in loop. Infligge danno continuo al player (con timer) quando resta a contatto. |
| **PhantomMover.cs** | Nemico "Fantasma" che si muove in direzioni casuali ogni 2 secondi. Reagisce ai proiettili ma non insegue il player. |

---

## Player_Scripts/

| File | Descrizione |
| --- | --- |
| **PlayerAnimationHandler.cs** | Gestisce le animazioni del player (movimento, danno, morte). Simile a `EnemiesAnimationHandler` ma per il personaggio giocabile. |
| **PlayerController.cs** | Controller principale del player. Legge input da tastiera (WASD/frecce) e passa i comandi di movimento a `TopDownMover2D`. |

---

## Utils/

| File | Descrizione |
| --- | --- |
| **AudioManager.cs** | Contenitore per `AudioSource` e `AudioClip` (background, passi, sparo). Al momento solo definizioni, nessuna logica implementata. |
| **Bullet.cs** | Proiettile con velocità, danno e durata. Supporta modalità homing (segue un target). Applica danno tramite `LifeController.TakeDamage()`. |
| **CameraFollow.cs** | Camera che segue il player (mantiene Z a -10 per la prospettiva 2D). |
| **LifeController.cs** | Sistema di vita generico (player/nemici). Gestisce HP, danno, cura e morte. Max HP limitato a 999. Disabilita controller e attiva animazione di morte. |
| **PlayerFootsteps.cs** | Gestione suoni dei passi (codice commentato). Struttura presente per riprodurre suoni basati sulla velocità del movimento. |
| **TopDownMover2D.cs** | Movimento top-down 2D via `Rigidbody2D`. Riceve direzione normalizzata e muove il corpo con `MovePosition()`. |

---

## Weapons/

| File | Descrizione |
| --- | --- |
| **PickUp.cs** | Oggetto raccoglibile. Quando il player entra nel trigger, aggiunge l'arma al `PlayerWeaponList` e si distrugge. |
| **PlayerWeaponList.cs** | Gestisce la raccolta armi. Se l'arma esiste già, la potenzia (`LevelUp`); altrimenti la istanzia e la aggiunge al `WeaponManager`. |
| **ShooterController.cs** | Controlla lo sparo automatico. Ogni frame chiama `Shoot` su tutte le armi. Determina la direzione in base al tipo di arma (mouse, fisso, ecc.). |
| **Weapon.cs** | Classe arma con nome, danno, fire rate, prefab proiettile. Supporta double shot e homing. Metodo `LevelUp()` per potenziamento. |
| **WeaponManager.cs** | Inventario armi del player. Gestisce aggiunta, potenziamento e rimozione armi distrutte. Emette evento `OnWeaponListChanged`. |
